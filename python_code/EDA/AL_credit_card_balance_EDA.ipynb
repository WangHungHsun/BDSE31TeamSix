# 欄位介紹看csv檔

```python
credit_card_balance.sort_values(by=['SK_ID_CURR','SK_ID_PREV'])
#384萬筆資料，23欄位
```

```python
# 總行數
row_total = credit_card_balance.shape[0]
```

```python
# 遺失值比率
# 全部遺失值不超過20% 
nan_rows_sum = credit_card_balance.isna().sum()/row_total
max_nan_percentage = nan_rows_sum.max()
min_nan_percentage = nan_rows_sum.min()
print("最大遺失值佔比:", max_nan_percentage)
print("-"*80)
print(nan_rows_sum)
```

```python
# 0的比率
zero_rows_sum = (credit_card_balance == 0).sum()/row_total
print(zero_rows_sum)
#print("-"*80)
#print(zero_rows_sum.describe())
```

```python
#有幾種類別
MONTHS_BALANCE_Sort = credit_card_balance.sort_values(by=['MONTHS_BALANCE'],ascending=False)
MONTHS_BALANCE_Sort_Categories = MONTHS_BALANCE_Sort['MONTHS_BALANCE'].unique()
print(MONTHS_BALANCE_Sort_Categories)
```

```python

# 看Active佔的總比數
Boolin = credit_card_balance['NAME_CONTRACT_STATUS']=='Active'
print(Boolin.sum()/row_total)
```

```python
# 發現描述裡"AMT_RECIVABLE"與"AMT_TOTAL_RECEIVABLE" 極度相似
matching_count = (credit_card_balance['AMT_RECIVABLE'] == credit_card_balance['AMT_TOTAL_RECEIVABLE']).sum()
print("相同的百分比:", matching_count/row_total)
print("選擇drop掉其一")
print("-"*80)
RE_NAN=credit_card_balance["AMT_RECIVABLE"].isna().sum()/row_total
RE_ZERO=(credit_card_balance["AMT_RECIVABLE"] == 0).sum()/row_total
RETOTAL_NAN=credit_card_balance["AMT_TOTAL_RECEIVABLE"].isna().sum()/row_total
RETOTAL_ZERO=(credit_card_balance["AMT_TOTAL_RECEIVABLE"] == 0).sum()/row_total
print("AMT_RECIVABLE 遺失值為 : ",RE_NAN)
print("AMT_TOTAL_RECEIVABLE 遺失值為 : ",RETOTAL_NAN)
print("-"*80)
print("AMT_RECIVABLE        0值為 : ",RE_ZERO)
print("AMT_TOTAL_RECEIVABLE 0值為 : ",RETOTAL_ZERO)
print("-"*80)
print("所以drop掉0值比較多的AMT_Recivable")
```
