```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import KNNImputer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import re
pd.options.display.float_format = '{:.7f}'.format
pd.set_option('display.max_columns', None)
```

```python
all_v1=pd.read_csv('D:/project/homecredit/final/v1/train_all_v1.csv', index_col=0)
all_v2 = pd.read_csv('D:/project/homecredit/final/v2/ETL_ALL_not_encoding_v2.csv', index_col=0)
```

```python
all_v1 = all_v1.iloc[:,38:]
```

```python
final_all_v2 = pd.concat([all_v2,all_v1],axis=1)
```

```python
for column in final_all_v2.columns:
    # 使用正则表达式将引号和特殊字符替换为下划线
    new_column = re.sub(r'["\'\W]', '_', column)
    
    # 使用rename方法为列设置新的列名
    final_all_v2.rename(columns={column: new_column}, inplace=True)

# 现在final_all_v3_new包含了新的列名
```

```python
# final_all_v2.to_csv('D:/project/homecredit/final/v2/final_all_v2.csv',index = False)
```

```python
final_all_v2_onehot = pd.get_dummies(final_all_v2, drop_first=True)
final_all_v2_onehot = final_all_v2_onehot.drop(columns=['CODE_GENDER_XNA'])
for column in final_all_v2_onehot.columns:
    column_type = type(final_all_v2_onehot[column][0])  # 获取第一行数据的类型
    print(f"列名: {column}，数据类型: {column_type}")

```

```python
column_types = final_all_v2_onehot.dtypes
uint8_columns = column_types[column_types == 'uint8'].index
final_all_v2_onehot[uint8_columns] = final_all_v2_onehot[uint8_columns].astype('int64')
```

```python
column_types = final_all_v2_onehot.dtypes

# 计算每种数据类型的列数
type_counts = column_types.value_counts()

print("各种数据类型的列数:")
print(type_counts)
```
