```python
import pandas as pd
import numpy as np

# Read the CSV file
file_path = '㊃test_pre_app_DAYS_EMPLOYED_filled.csv'
pre_app = pd.read_csv(file_path)

# (1) Replace values of 100000000 in 'AMT_INCOME_TOTAL' column with np.nan
pre_app['AMT_INCOME_TOTAL'] = pre_app['AMT_INCOME_TOTAL'].replace(100000000, np.nan)

# (2) Take the absolute value of 'DAYS_BIRTH' column
pre_app['DAYS_BIRTH'] = pre_app['DAYS_BIRTH'].abs()

# (3) Take the absolute value of 'DAYS_REGISTRATION' column
pre_app['DAYS_REGISTRATION'] = pre_app['DAYS_REGISTRATION'].abs()

# (4) Take the absolute value of 'DAYS_ID_PUBLISH' column
pre_app['DAYS_ID_PUBLISH'] = pre_app['DAYS_ID_PUBLISH'].abs()

# (5) Replace missing values in 'OWN_CAR_AGE' column with 0
pre_app['OWN_CAR_AGE'].fillna(0, inplace=True)

# (6) Replace missing values in 'OCCUPATION_TYPE' column with np.nan
pre_app['OCCUPATION_TYPE'].fillna(np.nan, inplace=True)

# (7) Replace values of -1 in 'REGION_RATING_CLIENT_W_CITY' column with 0
pre_app['REGION_RATING_CLIENT_W_CITY'] = pre_app['REGION_RATING_CLIENT_W_CITY'].replace(-1, 0)

# (8) Replace missing values in 'EXT_SOURCE_1', 'EXT_SOURCE_2', and 'EXT_SOURCE_3' columns with 0
pre_app[['EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3']] = pre_app[['EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3']].fillna(0)

# (9) Take the absolute value of 'DAYS_LAST_PHONE_CHANGE' column
pre_app['DAYS_LAST_PHONE_CHANGE'] = pre_app['DAYS_LAST_PHONE_CHANGE'].abs()

# (10) Replace missing values in 'AMT_REQ_CREDIT_BUREAU_MON' and 'AMT_REQ_CREDIT_BUREAU_YEAR' columns with 0
pre_app[['AMT_REQ_CREDIT_BUREAU_MON', 'AMT_REQ_CREDIT_BUREAU_YEAR']] = pre_app[['AMT_REQ_CREDIT_BUREAU_MON', 'AMT_REQ_CREDIT_BUREAU_YEAR']].fillna(0)

# (11) Replace values greater than 100 in 'AMT_REQ_CREDIT_BUREAU_QRT' column with np.nan
pre_app['AMT_REQ_CREDIT_BUREAU_QRT'] = pre_app['AMT_REQ_CREDIT_BUREAU_QRT'].apply(lambda x: np.nan if x > 100 else x)

# Export the modified DataFrame to a new CSV file
output_file_path = '㊄application_test_clean.csv'
pre_app.to_csv(output_file_path, index=False)

# Display the first few rows of the cleaned DataFrame
print(pre_app.head())
```

# ★合表(Test+Previous)

```python
import pandas as pd

# Read the CSV files
file_path_app = '㊅application_test_clean.csv'
file_path_prev = '⑪previous_data.csv'

app_data = pd.read_csv(file_path_app)
prev_data = pd.read_csv(file_path_prev)

# Merge the dataframes based on 'SK_ID_CURR'
merged_data = pd.merge(app_data, prev_data, on='SK_ID_CURR', how='inner')

# Display the first few rows of the merged DataFrame
print(merged_data.head())

# Export the merged DataFrame to a new CSV file
output_file_path = '⑴merged(Test&Previous_data.csv'
merged_data.to_csv(output_file_path, index=False)
```
