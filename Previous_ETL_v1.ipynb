```python
import pandas as pd

# 读取CSV文件
data = pd.read_csv('previous_application.csv')

# 使用pivot_table进行长表到宽表的转换
pivot_data = data.pivot_table(index=['SK_ID_CURR'], columns=['NAME_CONTRACT_TYPE'], aggfunc='size', fill_value=0)

# 重新设置列名
pivot_data.columns = ['NAME_CONTRACT_TYPE_' + col for col in pivot_data.columns]

# 重置索引
pivot_data.reset_index(inplace=True)

# 打印转换后的宽表头部结果
pivot_data.head()
```

# ⭐⭐⭐這以下+三個合表

```python
#⭐⭐⭐合表用
import pandas as pd

# 读取CSV文件
data = pd.read_csv('previous_application_dropped.csv')

# 使用groupby对数据进行分组，并计算平均值
grouped_data = data.groupby(['SK_ID_CURR', 'NAME_CONTRACT_TYPE']).mean()

# 重置索引，保留分组列作为索引的一部分
grouped_data.reset_index(inplace=True)

# 使用Pandas来筛选满足条件的行
filtered_data = grouped_data[grouped_data['NAME_CONTRACT_TYPE'] == 'Consumer loans']

# 重新命名列
filtered_data.rename(columns={'AMT_ANNUITY': 'AMT_ANNUITY_Consumer loans_Mean',
                              'AMT_APPLICATION': 'AMT_APPLICATION_Consumer loans_Mean',
                              'AMT_CREDIT': 'AMT_CREDIT_Consumer loans_Mean',
                              'AMT_GOODS_PRICE': 'AMT_GOODS_PRICE_Consumer loans_Mean'},
                     inplace=True)

# 打印满足条件的结果的头部
filtered_data
```
